```
import pandas as pd
import numpy as np 
import seaborn as sns

pd.options.display.float_format = '{:,.3f}'.format
```

```
df = sns.load_dataset('tips')
df.head()
```

```
df.describe(include='all')
```

```
df['day'].value_counts()
```

```
#porcentaje
df['day'].value_counts(normalize=True)*100
```

```
df.groupby('sex').mean()
```

```
df['pct_tip'] = df['tip']/df['total_bill']
df.head()
```

```
df.groupby('sex').mean()
```

```
df.groupby('sex').median()
```

```
df.groupby('sex')[['total_bill','pct_tip']].describe()
```

```
def mean_eur2eusd(x):
  return np.mean(x)*1.12
```

```
df.groupby('sex')[['total_bill','pct_tip']].apply(mean_eur2eusd)
```

```
df.groupby(['sex','time'])[['total_bill','pct_tip']].apply(mean_eur2eusd)
```

```
df.groupby(['sex','time'])[['total_bill','pct_tip']].apply(np.std)
```

```
df.groupby(['sex','time'])[['total_bill','pct_tip']].apply(lambda x: np.mean(x)*1.12)
```

```
#df.groupby(['sex','time'])[['total_bill','pct_tip']].aggregate([np.mean, np.max])
#usando agg en lugar de aggregate funciona igual
df.groupby(['sex','time'])[['total_bill','pct_tip']].agg([np.mean, np.max])
```

```
dict_agg = {'tip':[min,max], 'total_bill':[np.mean, mean_eur2eusd] }
```

```
df.groupby(['sex','time'])[['total_bill','tip']].agg(dict_agg)
```

```
def f_filter(x):
  return mean_eur2eusd(x['total_bill']) > 20
```

```
df.groupby(['sex','time']).filter(f_filter)
```

```
#Agregandolo a el df 
df['bin_total'] = pd.cut(df['total_bill'], bins= [3,18,35,60])
df.head()
```

```
df.groupby(['time','bin_total'])[['ones']].count()
```

```
df.groupby(['time','bin_total'])[['ones']].count().groupby(level=0).apply(
    lambda x:
    x / x.sum() * 100
)
```

```
df_pivot = df.pivot_table(values = 'total_bill', index = 'sex', columns='time', aggfunc=[np.median, np.std])
df_pivot
```
