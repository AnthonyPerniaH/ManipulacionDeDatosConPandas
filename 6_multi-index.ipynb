```
df = pd.read_csv('population.csv', skiprows=4)
df.tail()
```

```
df = df.drop(columns=['Country Code','Indicator Name','Indicator Code','Unnamed: 64'])

df = pd.melt(df, id_vars=['Country Name'],
var_name='year', value_name='pob')

df
```
