```
df_lmerged = pd.read_csv('london_merged.csv')
df_lmerged.head()
```

```
df_lmerged['timestamp'] = pd.to_datetime(df_lmerged['timestamp'])
```

```
df_lmerged.dtypes
```

```
df_lmerged['hour'] = df_lmerged['timestamp'].dt.hour
df_lmerged.head()
```

```
#quito el tiempo (la primera)
df = df_lmerged.iloc[: , 1:]
df.head()
```

```
#Al cuadrado
df['wind_speed']**2
```

```
#de 2 en dos 
df['t1'].iloc[::2]
```

```
#de 3 en 3
df['t1'].iloc[::3]
```

```
#El sub es una resta 20 - 10    , es igual 20 .sub(10, fill_value=999)  , en fill_value, va el dato para cuando encuentre datos NaN, al 999 se restaria

df['t1'].iloc[::3].sub(df['t2'], fill_value = 10000)
```

```
#producto punto
df['t1'].dot(df['t1'])
```

```
def func_1(x):
  y = x**2 + 1

  return y 
  
```

```
df['hour'].apply(func_1)
```

```
def func_2(x, a = 1, b = 0):
  y = x **2 + a*x + b

  return y 
```

```
#pasando argumentos version 1 
df['hour'].apply(func_2, args=(20, -100))
```

```
#pasando argumentos version 2 
df['hour'].apply(func_2, a= 20, b = -100)
```
